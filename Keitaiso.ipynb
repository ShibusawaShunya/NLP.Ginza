```python
#トークン化

import spacy 

nlp = spacy.load('ja_ginza')
doc = nlp('青い大きな目の猫')

for token in doc:
  print(token)
```

```python
#トークン化A

import ginza
import spacy

nlp = spacy.load('ja_ginza')
ginza.set_split_mode(nlp, 'A') #分割単位A = 最も短い言葉の単位
doc = nlp('私は国家公務員です')

for token in doc:
  print(token)
```

```python
#トークン化B

import ginza
import spacy

nlp = spacy.load('ja_ginza')
ginza.set_split_mode(nlp, 'B') #分割単位B = A単位＋接辞、および一般の複合動詞
doc = nlp('私は国家公務員です')

for token in doc:
  print(token)
```

```python
#トークン化C

import ginza
import spacy

nlp = spacy.load('ja_ginza')
ginza.set_split_mode(nlp, 'C') #分割単位C = 複合名詞、固有名詞、慣用句など
doc = nlp('私は国家公務員です')

for token in doc:
  print(token)
```

```python
#品詞タグ付け

import spacy

nlp = spacy.load('ja_ginza')
doc = nlp('銀座に行きます。')

for token in doc:
  print(
      token.text+', '+ #テキスト
      token.tag_+', '+ #SudachiPyの品詞タグ
      token.pos_)      #Universal Dependenciesの品詞タグ
```

```python
#レンマ化

import spacy

nlp = spacy.load('ja_ginza')
doc = nlp('銀座に行きます。')

for token in doc:
  print(
      token.text+', '+ #テキスト
      token.lemma_)    #レンマ化
```

```python
#文境界解析

import spacy

nlp = spacy.load('ja_ginza')
doc = nlp('銀座でランチをご一緒しましょう。今度の日曜日はどうですか。')

for span in doc.sents:
  print(span)
```

```python
#文境界解析＋トークン化

import spacy

nlp = spacy.load('ja_ginza')
doc = nlp('銀座でランチをご一緒しましょう。今度の日曜日はどうですか。')

for span in doc.sents:
  print(span)
  for token in span:
    print(token)
```

```python
#分節分割

import spacy
import ginza

nlp = spacy.load('ja_ginza')
doc = nlp('銀座でランチをご一緒しましょう。今度の日曜日はどうですか。')

for sent in doc.sents:
  for span in ginza.bunsetu_spans(sent):
    print(span)
```

```python
#分節分割＋トークン化

import spacy
import ginza

nlp = spacy.load('ja_ginza')
doc = nlp('銀座でランチをご一緒しましょう。今度の日曜日はどうですか。')

for sent in doc.sents:
  for span in ginza.bunsetu_spans(sent):
    for token in span:
      print(token)
```
